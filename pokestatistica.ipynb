### 3) Análise de Frequência dos TIPOS e STATS

```python
# quão comum é cada tipo primário?
tipo1_Fi = df['type1'].value_counts() # absoluta
tipo1_fi = df['type1'].value_counts(normalize=True)
tipo1_fai = tipo1_fi.cumsum()

tipo1_frequencias = pd.DataFrame({'Fi': tipo1_Fi,
                                  'fi': tipo1_fi,
                                  'fai': tipo1_fai})
```
